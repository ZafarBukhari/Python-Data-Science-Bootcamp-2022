# <font color="Green">2- How to Split Dataset into Two Sub-Datasets?</font>

```python
# This will again load a Dataset from Seaborn built-in dataset templates

titanic_df = sns.load_dataset('titanic')
titanic_df.head()
```

```python
# This will display total number of rows in this dataset

len(titanic_df)
```

```python
# We can also explore it with this alternative way
# This will display total number of columns and rows

titanic_df.shape
```

```python
# This will import 'random' module

from random import random
```

```python
# This will create a new dataset from the already created dataset 'titanic_df'
# This will be '0.5' fraction of 'titanic_df' dataset

boat_1 = titanic_df.sample(frac=0.5, random_state=1)
boat_1.shape 
```

```python
# Here, another dataset will be created by subtracting 'boat_1' dataset from the larger dataset 'titanic_df'
# This will disply (445, 15) i.e. Larger Dataset Columns - Boat_1 Columns ( 981 - 446 = 445)

boat_2 = titanic_df.drop(boat_1.index)
boat_2.shape
```

```python
# Now, we will display first five rows of each dataset

boat_1.head()
```

```python
# This will show rows other than of 'boat_1'

boat_2.head()
```

```python
# As 'boat_2' dataset is derived by subtracting 'boat_1' from 'titanic_df', so 'tatanic_df=boat_1 + boat_2'
# This will show number of rows equal to the number of rows of 'titanic_df'

len(boat_1) + len(boat_2)
```

**Note: We can also merge rows of both sub-datasets, as under: -**

# <font color="Green">3- How to Join Two Datasets?</font>

```python
# This will append 'boat_2' dataset at the end of 'baot_1' dataset

new_df = boat_1.append(boat_2)
new_df.shape
```

```python
# Now, we will display the columns list to further filter the dataset

titanic_df.columns
```

```python
# Now, we will do filtering based on column 'embark_town'
# First, we will display unique values of column 'embark_town'

titanic_df.embark_town.unique()
```

```python
# Now, we will display records of all those passengers who embarked from 'Queenstown'

titanic_df[(titanic_df.embark_town=='Queenstown')]
```

```python
# We can also display count of those records who embarked at 'titanic' ship from 'Queenstown'

titanic_df[(titanic_df.embark_town=='Queenstown')].shape
```

```python
# We can also filter the dataset based on multiple column lables
# This shows that there were 613 male passengers at the 'titanic' ship who were embarked from 'Queenstown'

titanic_df[(titanic_df.embark_town=='Queenstown') & (titanic_df.sex=='male' )].shape
```

```python
# This will display count of columns where embark_town = 'Cherbourg'

titanic_df[titanic_df.embark_town.isin(['Cherbourg'])].count()
```

```python
# We can also display count of columns where embark_town = 'Cherbourg' & 'Queenstown'

titanic_df[titanic_df.embark_town.isin(['Cherbourg','Queenstown'])].count()
```

```python
# We can also the records based on filtering 'age' column

titanic_df[titanic_df.age>30].shape
```

```python
# We can also the records based on filtering 'age' column as well as 'sex' column

titanic_df[(titanic_df.age>30) & (titanic_df.sex=='female')].shape
```

```python
titanic_df.alone.unique()
```

```python
# We can also the records based on criteria, as under: -
# Those passengers whose age is greater than '30' and whose gender is 'male' and also who were 'alone'

titanic_df[(titanic_df.age>30) & (titanic_df.sex=='male') & (titanic_df.alone==True)].shape
```

# <font color="Green">5- How to Filter by Large Categories?</font>

```python
# This will display the number of passengers boarded from each city in the 'titanic' ship

titanic_df.embark_town.value_counts()
```

```python
# This will display 'gender' wise number of passengers in the 'titanic' ship

titanic_df.sex.value_counts()
```

```python
# This will display 'age' wise count of passengers in the 'titanic' ship

titanic_df.age.value_counts()
```

**In the above output, we seen that total length of rows in 'age wise' output consists of 88 rows.<br> So, we will further filter it by *largest age parameter i.e. <font color="Red"> 'nlargest()'</font>, as under: -**

```python
# This will display 'age' wise count of passengers in the 'titanic' ship

titanic_df.age.value_counts().nlargest(3)
```

```python
# This will display index based on 'age' wise count of passengers in the 'titanic' ship

titanic_df.age.value_counts().nlargest(3).index
```

```python
# Another way to select the data by largest categories

counts = titanic_df.age.value_counts()
counts.nlargest(5)
```

```python
# This will display index of passengers with 'age' parameter

counts = titanic_df.embark_town.value_counts()
counts.nlargest().index
```

```python
# This will display count of passengers with 'embark_town' parameter

counts = titanic_df.embark_town.value_counts()
counts
```

```python
# This will display result of the largest ONE category i.e. which is of 'city' Southampton

titanic_df[titanic_df.embark_town.isin(counts.nlargest(1).index)].head()
```
